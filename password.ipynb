```python


password = input("Enter your password: ")


if len(password) != 12:
    print("Password must be exactly 12 characters long.")
else:
    p_upper = False
    p_digit = False
    p_special = False

    for char in password:
        ascii_val = ord(char)

        if 65 <= ascii_val <= 90:
            p_upper = True
        elif 48 <= ascii_val <= 57:
            p_digit = True
        elif (33 <= ascii_val <= 47) or (58 <= ascii_val <= 64) or (91 <= ascii_val <= 96) or (123 <= ascii_val <= 126):
            p_special = True

    if p_upper and p_digit and p_special:
        print("Password accepted.")
    else:
        print("your password is missing special character or digit or captial")
```

```python
def upper_case(user_input):
    for char in user_input:
        ascii_val = ord(char)
        if 65 <= ascii_val <= 90:
            return True
    return False

def is_digit_validator(user_input):
    for char in user_input:
        ascii_val = ord(char)
        if 48 <= ascii_val <= 57:
            return True
    return False

def is_character_validator(user_input):
    count = 0
    for char in user_input:
        ascii_val = ord(char)
        if (65 <= ascii_val <= 90) or (97 <= ascii_val <= 122):
            count += 1
    if count >= 5:
        return True
    return False

def is_special_character(user_input):
    for char in user_input:
        ascii_val = ord(char)
        if (33 <= ascii_val <= 47) or (58 <= ascii_val <= 64) or (91 <= ascii_val <= 96) or (123 <= ascii_val <= 126):
            return True
    return False

def length(user_input):
    if len (user_input) == 12:
        return True
    return False

def password_generator(user_input):
    if (upper_case(user_input) and length(user_input) and is_digit_validator(user_input)
        and is_character_validator(user_input) and is_special_character(user_input)):
        print("Your password is valid")
    else:
        print("Your password is not valid")


user_input = input("Enter your password: ")
password_generator(user_input)
```
