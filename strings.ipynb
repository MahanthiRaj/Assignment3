# ***Swapcase***
Makes uppercase letters lowercase and lowercase letters uppercase.

```python
a = "hello"
print(a.swapcase())
```

```python
a = "HELLO"
print(a.swapcase())
```

# ***lstrip and rstrip***

lstrip(): Removes spaces at the beginning of the string.


rstrip(): Removes spaces at the end of the string.



```python
print(my_string.lstrip())
print(my_string.rstrip())
```

# ***zfill***

The zfill() method pads a string with zeros (0) on the left to ensure that the string reaches a specified length.


```python
my_string = "42"
padded_string = my_string.zfill(5)
print(padded_string)  # Output: '00042'
```

# ***center***

The center() method centers a string within a
specified width by padding it with spaces


```python
my_string = "Python"
cente = my_string.center(20, "-")
print(cente)
```
