```python
import random
```

```python

u1_total = 0
u2_total = 0
limit = 20

for turn in range(100):
    u1_roll = random.randint(1, 6)
    u1_total += u1_roll
    print("u1 rolled" ,u1_roll,  "Total:",u1_total)
    if u1_total >= limit:
        print("u1 won the game!")
        break

    u2_roll = random.randint(1, 6)
    u2_total += u2_roll
    print("u2 rolled:" ,u1_roll,  "Total:",u2_total)
    if u2_total >= limit:
        print("u2 won the game!")
        break

```

```python
u1_input = input("Enter Player 1's name: ")
u2_input = input("Enter Player 2's name: ")

def dice_game(u1_input, u2_input):
    u1_total = 0
    u2_total = 0
    limit=20

    for turn in range(1, 21):
        input("Press any key to roll the dice for Player 1: ")
        u1_roll = random.randint(1, 6)
        u1_total += u1_roll
        print(f"Turn {turn}: u1 rolled {u1_roll}, Total: {u1_total}")
        if u1_total >= limit:
            print("u1 won the game!")
            break

        input("Press any key to roll the dice for Player 2: ")
        u2_roll = random.randint(1, 6)
        u2_total += u2_roll
        print(f"Turn {turn}: u2 rolled {u2_roll}, Total: {u2_total}")
        if u2_total >= limit:
            print("u2 won the game!")
            break

dice_game(u1_input, u2_input)


```
